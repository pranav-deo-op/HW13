```python
# Import necessary libraries
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd

# Load the iris dataset
iris = datasets.load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# Line Plot
plt.figure(figsize=(10, 6))
for species in iris_df['species'].unique():
    subset = iris_df[iris_df['species'] == species]
    plt.plot(subset['sepal length (cm)'].values, label=f'Species {species}')
plt.title('Line Plot of Sepal Length for Each Species')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.legend()
plt.show()

# Bar Chart
mean_sepal_length = iris_df.groupby('species')['sepal length (cm)'].mean()
plt.figure(figsize=(10, 6))
plt.bar(mean_sepal_length.index, mean_sepal_length.values, tick_label=[f'Species {i}' for i in mean_sepal_length.index])
plt.title('Average Sepal Length for Each Species')
plt.xlabel('Species')
plt.ylabel('Average Sepal Length (cm)')
plt.show()

# Histogram
plt.figure(figsize=(10, 6))
plt.hist(iris_df['sepal length (cm)'], bins=20, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()

# Box Plot
plt.figure(figsize=(10, 6))
plt.boxplot(iris_df['sepal length (cm)'], notch=True, vert=False)
plt.title('Box Plot of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.show()

# Scatter Plot
plt.figure(figsize=(10, 6))
for species in iris_df['species'].unique():
    subset = iris_df[iris_df['species'] == species]
    plt.scatter(subset['sepal length (cm)'], subset['sepal width (cm)'], label=f'Species {species}')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.show()
```
